```python
# Q1. How many Customers do we have in the data?

query = """
SELECT COUNT(DISTINCT(customer_id)) AS customers
FROM customers
"""
sqldf(query)
```

```python
# Q2. What was the city with the most profit for the company in 2015 and how much was it?

query = """
SELECT
o.shipping_city AS city,
o.shipping_country AS country,
SUM(od.order_profits) AS total_profit,
CASE WHEN o.shipping_country = 'United States' THEN 'USD' ELSE 'Not_USD' END AS currency
FROM orders AS o
JOIN order_details AS od
ON o.order_id = od.order_id
WHERE order_date LIKE '%2015'
GROUP BY 1
ORDER BY 3 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q3. How many different cities do we have in the data?

query = """
SELECT
COUNT(DISTINCT(shipping_city)) AS city_count
FROM orders
"""
sqldf(query)
```

```python
# Q4. Show the total spent by customers from low to high.

query = """
SELECT
c.customer_id AS customer_id,
c.customer_name AS customer_name,
SUM(od.order_sales) AS total_spent,
CASE WHEN o.shipping_country = 'United States' THEN 'USD' ELSE 'Not_USD' END AS currency
FROM customers AS c
JOIN orders AS o
ON c.customer_id = o.customer_id
JOIN order_details AS od
ON o.order_id = od.order_id
GROUP BY 2
ORDER BY 3 ASC
"""
sqldf(query)
```

```python
# Q5. What is the most profitable City in the State of Tennessee?

query = """
SELECT
o.shipping_city AS city,
o.shipping_state AS state,
SUM(od.order_profits) AS total_profits,
CASE WHEN o.shipping_country = 'United States' THEN 'USD' ELSE 'Not_USD' END AS currency
FROM customers AS c
JOIN orders AS o
ON c.customer_id = o.customer_id
JOIN order_details AS od
ON o.order_id = od.order_id
WHERE o.shipping_state = 'Tennessee'
GROUP BY 1
ORDER BY 3 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q6. What’s the average annual profit for that city across all years in that city?

query = """
SELECT
o.shipping_city AS city,
AVG(od.order_profits) AS average_profits,
CASE WHEN o.shipping_country = 'United States' THEN 'USD' ELSE 'Not_USD' END AS currency
FROM customers AS c
JOIN orders AS o
ON c.customer_id = o.customer_id
JOIN order_details AS od
ON o.order_id = od.order_id
WHERE o.shipping_city = 'Lebanon'
"""
sqldf(query)
```

```python
# Q7. What is the distribution of customer types in the data?

query = """
SELECT
customer_segment,
COUNT(customer_segment) AS count
FROM customers
GROUP BY 1
"""
sqldf(query)
```

```python
# Q8. What’s the most profitable product category on average in Iowa across all years?

query = """
SELECT
product_category,
shipping_state,
AVG(order_profits) AS average_profit,
CASE WHEN o.shipping_country = 'United States'
THEN 'USD' ELSE 'Not_USD' END AS currency
FROM product AS p
JOIN order_details AS od
ON p.product_id = od.product_id
JOIN orders AS o
ON od.order_id = o.order_id
WHERE shipping_state = 'Iowa'
GROUP BY 1
ORDER BY 3 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q9. What is the most popular product in that category across all states in 2016?

# For this question, there were 11 popular products in the Furniture category across all states all tied 
# with a count of 4 orders placed by different customers in 2016. # So the first result with product names 
# and their number of orders was used to create a Common Table Expression (CTE) a.k.a subquery
# Then the qty of products per item was used instead to distinguish the most popular product amongst them.

query = """

WITH table1 AS (
    SELECT
        p.product_name,
        p.product_category,
        COUNT(*) AS order_count,
        SUBSTR(o.order_date, -4, 4)AS year
    FROM product AS p
    JOIN orders AS o
        ON o.order_id = od.order_id
    JOIN order_details AS od
        ON p.product_id = od.product_id
    WHERE product_category = 'Furniture' AND year = '2016'
    GROUP BY 1
    ORDER BY 3 DESC
    ),
    
table2 AS (
    SELECT product_name
    FROM table1
    WHERE order_count >=4
    )

SELECT product_name,
    product_category,
    COUNT(*) AS order_count,
    SUM(quantity) AS total_qty_ordered,
    SUBSTR(o.order_date, -4, 4)AS year
FROM product AS p
JOIN order_details AS od
    ON od.product_id = p.product_id
JOIN orders AS o
    ON o.order_id = od.order_id
WHERE product_name IN table2 AND year = '2016'
GROUP BY 1
ORDER BY 4 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q10. Which customer got the most discount in the data? (in total amount)

query = """
SELECT
c.customer_id AS customer_id,
c.customer_name AS customer_name,
SUM(ROUND((od.order_sales * od.order_discount)/(1-od.order_discount), 2)) AS total_discount_amount
FROM customers AS c
JOIN orders AS o
ON c.customer_id = o.customer_id
JOIN order_details AS od
ON o.order_id = od.order_id
GROUP BY 2
ORDER BY 3 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q11. How widely did monthly profits vary in 2018?

query = """

WITH table1 AS (
    SELECT
        SUM(od.order_profits) AS total_profit,
        CAST(SUBSTR(o.order_date, 1, instr(o.order_date, '/')-1) AS INT) AS month,
        SUBSTR(o.order_date, -4, 4)AS year
    FROM customers AS c
    JOIN orders AS o
        ON c.customer_id = o.customer_id
    JOIN order_details AS od
        ON o.order_id = od.order_id
    JOIN product AS p
        ON od.product_id = p.product_id
    WHERE SUBSTR(o.order_date, -4) = '2018'
    GROUP BY 2
    ORDER BY 2 ASC
    ),
    
table2 AS (
    SELECT
        month,
        total_profit,
        LAG(total_profit) OVER (ORDER BY month) AS running_total
    FROM table1
    )
    
SELECT
    month,
    total_profit,
    running_total,
    total_profit - running_total AS difference
FROM table2
ORDER BY 4 DESC
"""
print(sqldf(query))

print(f"\nHighest increase or decrease? -13824 USD in the month of 4 (April)")
```

```python
# Q12. Which order was the highest in 2015?

query = """
SELECT
od.order_id,
SUBSTR(o.order_date, -4, 4) AS year,
SUM(od.order_sales) AS total_order_sales
FROM orders AS o
JOIN order_details AS od
ON o.order_id = od.order_id
WHERE SUBSTR(o.order_date, -4, 4) = '2015'
GROUP BY 1
ORDER BY 3 DESC
LIMIT 1
"""
sqldf(query)
```

```python
# Q13. What was the rank of each city in the East region in 2015?

query = """

WITH table1 AS (
    SELECT
        o.shipping_city,
        COUNT(*) AS number_count
    FROM orders AS o
    JOIN order_details AS od
        ON o.order_id = od.order_id
    JOIN product AS p
        ON od.product_id = p.product_id
    WHERE SUBSTR(o.order_date, -4, 4) = '2015'
        AND shipping_region = 'East'
    GROUP BY 1
    ORDER BY 2 DESC
    )
    
SELECT shipping_city,
        number_count,
        RANK() OVER(ORDER BY number_count DESC) AS RANK
FROM table1
"""
print(sqldf(query))

print(f"\n{sqldf(query)['shipping_city'][0]} ranked 1st, while {sqldf(query)['shipping_city'][2]} came 3rd, with other cities being ranked with same rank for tied values")
```

```python
# Join all DB tables into one dataset that includes all unique columns and call it dataset. Then show the content of this one table to make sure all individual tables were joined.

query = """
SELECT
c.customer_id,
c.customer_name,
c.customer_segment,
o.order_id,
o.order_date,
SUBSTR(o.order_date, -4, 4)AS order_year,
CAST(SUBSTR(o.order_date, 1, instr(o.order_date, '/')-1) AS INT) AS order_month,
CAST(SUBSTR(o.order_date, (instr(o.order_date, '/')+1), (length(o.order_date) - instr(o.order_date, '/')-5)) AS INT) AS order_day,
o.shipping_city,
o.shipping_state,
o.shipping_region,
o.shipping_country,
o.shipping_postal_code,
o.shipping_date,
SUBSTR(o.shipping_date, -4, 4)AS shipping_year,
CAST(SUBSTR(o.shipping_date, 1, instr(o.shipping_date, '/')-1) AS INT) AS shipping_month,
CAST(SUBSTR(o.shipping_date, (instr(o.shipping_date, '/')+1), (length(o.shipping_date) - instr(o.shipping_date, '/')-5)) AS INT) AS shipping_day,
o.shipping_mode,
od.order_details_id,
od.quantity,
od.order_discount,
od.order_profits,
od.order_profit_ratio,
od.order_sales,
p.product_id,
p.product_name,
p.product_category,
p.product_subcategory,
p.product_manufacturer
FROM customers AS c
JOIN orders AS o
ON c.customer_id = o.customer_id
JOIN order_details AS od
ON o.order_id = od.order_id
JOIN product AS p
ON od.product_id = p.product_id
"""

dataset = sqldf(query)

# Viewing all information about the resulting DataFrame including the index dtype and columns, non-null values and memory usage.
dataset.info()
```
